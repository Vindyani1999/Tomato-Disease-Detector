```python
#all those heardcoded stuffs, we are going to define a fucntion

def get_dataset_partitions_tf(ds, train_split=0.8, validation_split=0.1, test_split=0.1, shuffle = True, shuffle_size = 1000):
    
    if shuffle:
        ds = ds.shuffle(shuffle_size, seed=12)
    
    ds_size = len(ds)
    train_size = int(train_split * ds_size)
    validation_size = int(validation_split * ds_size)
    
    train_ds = ds.take(train_size)
    
    validation_ds = ds.skip(train_size).take(validation_size)
    test_ds = ds.skip(train_size).skip(validation_size)
    
    return train_ds, test_ds, validation_ds
```

```python
train_ds, validation_ds, test_ds = get_dataset_partitions_tf(dataset)
```
